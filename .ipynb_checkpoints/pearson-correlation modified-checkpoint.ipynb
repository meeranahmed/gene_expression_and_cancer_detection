```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

```python
healthy_df = pd.read_csv("lusc-rsem-fpkm-tcga_paired.txt", sep = '\t')
```

```python
healthy_df.head()
```

```python
cancer_df = pd.read_csv("lusc-rsem-fpkm-tcga-t_paired.txt", sep = '\t')
```

```python
cancer_df.dropna()
```

```python
cancer_df.head()
```

```python
healthy_df.shape, cancer_df.shape
```

```python
ZERO_THRESHOLD = 30
```

```python
def contain_zeros(h_series, c_series):
    """
    contain_zeros: a function that counts the zeros in both cancer and healthy series and
    returns True if either exceeds ZERO_THRESHOLD, False otherwise
    """
    h_zeros = 0
    c_zeros = 0
    h_value_counts = h_series.value_counts()
    c_value_counts = c_series.value_counts()
    if 0 in h_value_counts.index:
        h_zeros = h_value_counts[0]
    else:
        h_zeros = 0
    if 0 in c_value_counts.index:
        c_zeros = c_value_counts[0]
    else:
        c_zeros = 0
        
    return (h_zeros > ZERO_THRESHOLD or c_zeros > ZERO_THRESHOLD)
```

```python
pearson_coef_dict = {}
```

```python
for i in range(healthy_df.shape[0]):
    gene_type = healthy_df.iloc[i, 0]
    healthy_values = healthy_df.iloc[i, 2:]
    cancer_values = cancer_df.iloc[i, 2:]
    zeros_status = contain_zeros(healthy_values, cancer_values)
    if not zeros_status: # if the number of zeros in both series doesn't exceed 30
        pearson_coef = healthy_values.astype(np.float64).corr(cancer_values.astype(np.float64)) 
        if not math.isnan(pearson_coef): # to ensure that correlation is not null-valued
            pearson_coef_dict[gene_type] = pearson_coef
```

```python
sorted_pearson_coef_dict = dict(sorted(pearson_coef_dict.items(), key = lambda pair: pair[1]))
for gene, pearson_coef in sorted_pearson_coef_dict.items():
    print(gene, pearson_coef)
```

```python
gene_lowest = list(sorted_pearson_coef_dict.items())[0][0]
gene_highest = list(sorted_pearson_coef_dict.items())[-1][0]
gene_lowest, gene_highest
```

```python
gene_lowest_healthy = healthy_df[healthy_df['Hugo_Symbol'] == gene_lowest].iloc[0, 2:].astype(np.float64)
gene_lowest_cancer = cancer_df[cancer_df['Hugo_Symbol'] == gene_lowest].iloc[0, 2:].astype(np.float64)
gene_highest_healthy = healthy_df[healthy_df['Hugo_Symbol'] == gene_highest].iloc[0, 2:].astype(np.float64)
gene_highest_cancer = cancer_df[cancer_df['Hugo_Symbol'] == gene_highest].iloc[0, 2:].astype(np.float64)
```

```python
plt.figure(figsize = (18, 10))

plt.subplot(1, 2, 1)
plt.title(gene_lowest)
sns.regplot(x = gene_lowest_healthy, y = gene_lowest_cancer)
plt.xlabel("Gene Expression Level - Healthy")
plt.ylabel("Gene Expression Level - Cancer")

plt.subplot(1, 2, 2)
plt.title(gene_highest)
sns.regplot(x = gene_highest_healthy, y = gene_highest_cancer)


plt.xlabel("Gene Expression Level - Healthy")
plt.ylabel("Gene Expression Level - Cancer");
```
